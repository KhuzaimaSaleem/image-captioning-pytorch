```python
import re

def clean_caption(caption):
    # Lowercase everything
    caption = caption.lower()
    # Remove special characters, keep only letters and spaces
    caption = re.sub(r"[^a-z\s]", "", caption)
    # Remove extra spaces
    caption = re.sub(r"\s+", " ", caption).strip()
    # Add start and end tokens
    caption = "<START> " + caption + " <END>"
    return caption

# Apply cleaning to all captions
df["caption"] = df["caption"].apply(clean_caption)

print("Cleaned captions sample:")
for cap in df["caption"].values[:5]:
    print(cap)
```

```python
from collections import Counter

# Collect all words from all captions (excluding special tokens)
all_captions = df["caption"].tolist()
all_words = " ".join(all_captions).split()

# Remove special tokens from word count
all_words = [w for w in all_words if w not in ["<START>", "<END>", "<PAD>", "<UNK>"]]

# Count word frequencies
word_counts = Counter(all_words)
print(f"Total unique words: {len(word_counts)}")

# Keep only words appearing 5 or more times
freq_words = [word for word, count in word_counts.items() if count >= 5]

# Special tokens FIRST, then rest of vocab
special_tokens = ["<PAD>", "<START>", "<END>", "<UNK>"]
vocab = special_tokens + freq_words

# word → index
word2idx = {word: idx for idx, word in enumerate(vocab)}

# index → word
idx2word = {idx: word for word, idx in word2idx.items()}

print(f"Vocabulary size: {len(vocab)}")
print("Index of <PAD>:", word2idx["<PAD>"])
print("Index of <START>:", word2idx["<START>"])
print("Index of <END>:", word2idx["<END>"])
print("Index of <UNK>:", word2idx["<UNK>"])
print(f"Sample: word 'dog' → index {word2idx.get('dog', 'not in vocab')}")
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import os

class FlickrDataset(Dataset):
    def __init__(self, df, img_dir, word2idx, transform=None):
        self.df = df
        self.img_dir = img_dir
        self.word2idx = word2idx
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        # Load image
        img_path = os.path.join(self.img_dir, row["image"])
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)

        # Convert caption to indices
        caption = row["caption"].split()
        caption_idx = [self.word2idx.get(word, self.word2idx["<UNK>"]) for word in caption]
        caption_tensor = torch.tensor(caption_idx, dtype=torch.long)

        return image, caption_tensor
```

```python
# Resize and normalize images for ResNet
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

# Create dataset
dataset = FlickrDataset(
    df=df,
    img_dir="/content/flickr8k/Images",
    word2idx=word2idx,
    transform=transform
)

print(f"Dataset size: {len(dataset)}")

# Test one sample
image, caption = dataset[0]
print(f"Image tensor shape: {image.shape}")
print(f"Caption tensor: {caption}")
print(f"Caption words: {[idx2word[idx.item()] for idx in caption]}")
```

```python
from torch.nn.utils.rnn import pad_sequence

def collate_fn(batch):
    images, captions = zip(*batch)

    # Stack all images into one tensor
    images = torch.stack(images, 0)

    # Pad captions to same length with <PAD> token (index 0)
    captions = pad_sequence(captions, batch_first=True, padding_value=0)

    return images, captions

# Create DataLoader
dataloader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=True,
    collate_fn=collate_fn,
    num_workers=2,
    pin_memory=True
)

print(f"Total batches: {len(dataloader)}")

# Test one batch
images, captions = next(iter(dataloader))
print(f"Batch image shape: {images.shape}")
print(f"Batch caption shape: {captions.shape}")
```

```python
import torch
import torch.nn as nn
import torchvision.models as models

class CNNEncoder(nn.Module):
    def __init__(self, embed_size):
        super(CNNEncoder, self).__init__()

        # Load pretrained ResNet50
        resnet = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)

        # Remove the last classification layer
        modules = list(resnet.children())[:-1]
        self.resnet = nn.Sequential(*modules)

        # Add our own linear layer to output embed_size features
        self.linear = nn.Linear(resnet.fc.in_features, embed_size)
        self.dropout = nn.Dropout(0.5)

        # Freeze ResNet weights — we don't want to retrain it
        for param in self.resnet.parameters():
            param.requires_grad = False

    def forward(self, images):
        # Extract features from images
        with torch.no_grad():
            features = self.resnet(images)

        # Flatten: [batch, 2048, 1, 1] → [batch, 2048]
        features = features.reshape(features.size(0), -1)

        # Pass through our linear layer → [batch, embed_size]
        features = self.dropout(self.linear(features))

        return features

# Test it
EMBED_SIZE = 256
encoder = CNNEncoder(embed_size=EMBED_SIZE)
print("CNN Encoder built ✅")
print(f"Output embed size: {EMBED_SIZE}")

# Quick test
dummy_images = torch.randn(4, 3, 224, 224)
output = encoder(dummy_images)
print(f"Encoder output shape: {output.shape}")  # Should be [4, 256]
```

```python
class TransformerDecoder(nn.Module):
    def __init__(self, embed_size, vocab_size, num_heads, num_layers, dropout=0.5):
        super(TransformerDecoder, self).__init__()

        # Convert word indices to vectors
        self.embedding = nn.Embedding(vocab_size, embed_size, padding_idx=0)

        # Positional encoding — tells model the order of words
        self.positional_encoding = nn.Embedding(100, embed_size)

        # Transformer decoder layers
        decoder_layer = nn.TransformerDecoderLayer(
            d_model=embed_size,
            nhead=num_heads,
            dim_feedforward=512,
            dropout=dropout,
            batch_first=True
        )
        self.transformer_decoder = nn.TransformerDecoder(
            decoder_layer,
            num_layers=num_layers
        )

        # Final layer — converts output to vocabulary scores
        self.fc_out = nn.Linear(embed_size, vocab_size)
        self.dropout = nn.Dropout(dropout)

    def forward(self, features, captions):
        batch_size, seq_len = captions.shape

        # Create position indices [0, 1, 2, ..., seq_len]
        positions = torch.arange(seq_len, device=captions.device).unsqueeze(0).expand(batch_size, -1)

        # Embed words + add positional encoding
        caption_embed = self.dropout(self.embedding(captions) + self.positional_encoding(positions))

        # Create causal mask — prevents model from cheating by looking ahead
        tgt_mask = nn.Transformer.generate_square_subsequent_mask(seq_len, device=captions.device)

        # Reshape features to [batch, 1, embed_size] as memory for decoder
        memory = features.unsqueeze(1)

        # Pass through transformer decoder
        output = self.transformer_decoder(
            tgt=caption_embed,
            memory=memory,
            tgt_mask=tgt_mask
        )

        # Convert to vocabulary scores
        output = self.fc_out(output)

        return output

# Test it
VOCAB_SIZE = len(vocab)
NUM_HEADS = 8
NUM_LAYERS = 3

decoder = TransformerDecoder(
    embed_size=EMBED_SIZE,
    vocab_size=VOCAB_SIZE,
    num_heads=NUM_HEADS,
    num_layers=NUM_LAYERS
)

print("Transformer Decoder built ✅")
print(f"Vocab size: {VOCAB_SIZE}")

# Quick test
dummy_features = torch.randn(4, 256)
dummy_captions = torch.randint(0, VOCAB_SIZE, (4, 10))
output = decoder(dummy_features, dummy_captions)
print(f"Decoder output shape: {output.shape}")  # Should be [4, 10, vocab_size]
```

```python
class ImageCaptioningModel(nn.Module):
    def __init__(self, embed_size, vocab_size, num_heads, num_layers, dropout=0.5):
        super(ImageCaptioningModel, self).__init__()
        self.encoder = CNNEncoder(embed_size)
        self.decoder = TransformerDecoder(embed_size, vocab_size, num_heads, num_layers, dropout)

    def forward(self, images, captions):
        # Step 1: Extract image features
        features = self.encoder(images)

        # Step 2: Generate caption scores
        output = self.decoder(features, captions)

        return output

# Build the full model
model = ImageCaptioningModel(
    embed_size=EMBED_SIZE,
    vocab_size=VOCAB_SIZE,
    num_heads=NUM_HEADS,
    num_layers=NUM_LAYERS
)

# Move model to GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

print(f"Model built ✅")
print(f"Training on: {device}")

# Count trainable parameters
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Trainable parameters: {total_params:,}")
```

```python
import torch.optim as optim

# Loss function — ignores <PAD> tokens (index 0)
criterion = nn.CrossEntropyLoss(ignore_index=0)

# Optimizer
optimizer = optim.Adam(model.parameters(), lr=3e-4)

# Learning rate scheduler — reduces LR when training plateaus
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode="min",
    patience=2,
    factor=0.5,
    verbose=True
)

print("Loss function ready ✅")
print("Optimizer ready ✅")
print("Scheduler ready ✅")
```

```python
import time

def train_one_epoch(model, dataloader, optimizer, criterion, device):
    model.train()
    total_loss = 0

    for batch_idx, (images, captions) in enumerate(dataloader):
        # Move data to GPU
        images = images.to(device)
        captions = captions.to(device)

        # Input to decoder: everything except last word
        captions_input = captions[:, :-1]

        # Target: everything except first word (<START>)
        captions_target = captions[:, 1:]

        # Forward pass
        optimizer.zero_grad()
        output = model(images, captions_input)

        # Reshape for loss calculation
        output = output.reshape(-1, VOCAB_SIZE)
        captions_target = captions_target.reshape(-1)

        # Calculate loss
        loss = criterion(output, captions_target)

        # Backward pass
        loss.backward()

        # Clip gradients to prevent exploding gradients
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

        # Update weights
        optimizer.step()
        total_loss += loss.item()

        # Print progress every 100 batches
        if (batch_idx + 1) % 100 == 0:
            print(f"  Batch {batch_idx+1}/{len(dataloader)} | Loss: {loss.item():.4f}")

    return total_loss / len(dataloader)

# ---- Training ----
NUM_EPOCHS = 10
best_loss = float("inf")

print("🚀 Starting Training!")
print(f"Training for {NUM_EPOCHS} epochs\n")

for epoch in range(NUM_EPOCHS):
    start_time = time.time()
    print(f"Epoch {epoch+1}/{NUM_EPOCHS}")

    train_loss = train_one_epoch(model, dataloader, optimizer, criterion, device)

    scheduler.step(train_loss)

    if train_loss < best_loss:
        best_loss = train_loss
        torch.save(model.state_dict(), "/content/best_model.pth")
        print(f"  ✅ Best model saved!")

    elapsed = time.time() - start_time
    print(f"  Epoch {epoch+1} done | Loss: {train_loss:.4f} | Time: {elapsed/60:.1f} mins\n")

print("🎉 Training Complete!")
```

```python
def generate_caption(model, image_tensor, word2idx, idx2word, device, max_len=30):
    model.eval()

    # Move image to GPU
    image_tensor = image_tensor.unsqueeze(0).to(device)

    # Encode image
    with torch.no_grad():
        features = model.encoder(image_tensor)

    # Start with <START> token
    caption = [word2idx["<START>"]]

    for _ in range(max_len):
        # Convert current caption to tensor
        cap_tensor = torch.tensor(caption, dtype=torch.long).unsqueeze(0).to(device)

        # Generate next word
        with torch.no_grad():
            output = model.decoder(features, cap_tensor)

        # Pick the word with highest score
        next_word_idx = output[0, -1, :].argmax().item()
        caption.append(next_word_idx)

        # Stop if <END> token is generated
        if next_word_idx == word2idx["<END>"]:
            break

    # Convert indices back to words
    words = [idx2word[idx] for idx in caption]
    words = [w for w in words if w not in ["<START>", "<END>", "<PAD>"]]

    return " ".join(words)


def show_image_with_caption(img_path, model, word2idx, idx2word, device, transform):
    # Load and show image
    image = Image.open(img_path).convert("RGB")
    plt.imshow(image)
    plt.axis("off")

    # Generate caption
    image_tensor = transform(image)
    caption = generate_caption(model, image_tensor, word2idx, idx2word, device)

    plt.title(f"Caption: {caption}", fontsize=12, wrap=True)
    plt.show()
    print(f"Generated Caption: {caption}")

# Load best model
model.load_state_dict(torch.load("/content/best_model.pth"))
print("Best model loaded ✅")

# Test on 5 random images
import random
sample_images = random.sample(os.listdir("/content/flickr8k/Images"), 5)

for img_name in sample_images:
    img_path = f"/content/flickr8k/Images/{img_name}"
    show_image_with_caption(img_path, model, word2idx, idx2word, device, transform)
    print("-" * 50)
```
